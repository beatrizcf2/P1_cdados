## 1.INTRODUÇÃO

A União Europeia, também conhecida como EU, é composta por 28 países, todos pertencentes ao continente europeu. Tal organização tem como objetivo principal a integração política e econômica entre os países europeus. Entre as políticas econômicas adotadas são:
* a livre circulação de pessoas e mercadorias
* a adoção de uma moeda única
* o mercado único europeu
* diminuição de taxas e impostos entre os países

## 2.FORMAÇÃO DA UNIÃO EUROPEIA

   Apesar de ter sido constituída apenas em 1992, sua formação vem desde o período pós segunda guerra mundial, em meados de 1952, quando alguns países europeus, dentre eles Alemanha, Itália, França, Bélgica, Holanda e Luxemburgo, uniram-se a partir da criação da Comunidade Europeia do Carvão e do Aço a fim de criar uma cooperação mútua, integrando o mercado siderúrgico.
   
   Posteriormente foi criado o Mercado comum europeu, em 1957, que contou com a participação de mais seis países: Inglaterra, Irlanda, Dinamarca, Grécia, Espanha e Portugal. A proposta dessa organização era de estabelecer a livre circulação de pessoas, capitais e serviços entre os países integrantes, característica que foi herdada pela atual união europeia. 
   
   Foi só em 1991 que a união europeia foi criada, a partir do Tratado de Maastricht.
