```python
def crescimento_anual(linhas):
    lista_crescimentos = []
    i=0
    while i<len(linhas)-1:
        co = linhas[i]
        c = linhas[i+1]
        y = (c-co)/co
        i+=1
        lista_crescimentos.append(y)
    return lista_crescimentos

print(crescimento_anual(linhas))

        

    
```
